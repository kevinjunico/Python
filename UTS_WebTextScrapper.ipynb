```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
# link used for this : https://en.wikipedia.org/wiki/Special:Search?search=sains+data&go=Go&ns0=1
```

```python
# scrap an html page
def get_text(url):
  req = requests.get(url)
  cont = req.content
  soup = bs(cont, 'html.parser')
  text = soup.get_text()
  return text
```

```python
# Make new text file and save list into text file
def new_txt(name,lst):
  with open(name+'.txt', 'w+') as f:
    for a in lst:
        f.write(a)
  f.close()
```

```python
# open file
def open_txt(name):
  with open(name+'.txt') as f:
    contents = f.read()
  return(contents)
```

```python
# Strip all newline
def strip_newLine(lst):
  strip = lst.replace('\n','')
  return(strip)
```

```python
# search unique words
def uniq_word(lst):
  lst = lst.lower()
  lists = lst.split()
  lists = [a.strip('.,!;()[]') for a in lists]
  lists = [a.replace("'s", '') for a in lists]

  unique = []
  unique_words = set(lists)
  for word in unique_words:
    unique.append(word)
  unique.sort()
  return(unique)
```

```python
# menu
def menu():
  print("Web Scrapper Menu")
  print("=================")
  print("1. Input web url")
  print("2. Show scrapping result")
  print("3. Save to .txt file")
  print("4. Open .txt file")
  print("5. Delete space")
  print("6. Gather unique words")
  print("9. Exit")
  print("=================")
```

```python
if __name__=='__main__':
  while(True):
    menu()
    choice = ''
    try:
      choice = int(input('Choose menu: '))
    except:
      print('Wrong input. Please enter a menu number ...')
    
    # Check choice and conditon
    if choice == 1:
      url = input("Enter web url : ")
      text = get_text(url)
      print("\nWeb Scrapping Finished")
    elif choice == 2:
      print("Showing the result...\n")
      print(text)
    elif choice == 3:
      name = input("Enter file name : ")
      new_txt(name, text)
      print("\nFile saved...")
    elif choice == 4:
      name = input("Enter file name to open : ")
      print("Opening file...\n")
      print(open_txt(name))
    elif choice == 5:
      text = strip_newLine(text)
      print(text)
      print("\nFinished proces...")
    elif choice == 6:
      unique = uniq_word(text)
      for i in unique:
        print(i)
    elif choice == 9:
      print('Thanks You!')
      break
    else:
      print("Menu Not Found!! Please enter a menu numbeer ...")
    
```
